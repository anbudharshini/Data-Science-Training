```python
import pandas as pd

data={
    'Name': ['Amit', 'Priya', 'Kiran', 'Anita', 'Ravi'],
    'Math': [78, 92, 65, 88, 55],
    'Science': [85, 89, 70, 95, 60],
    'English': [74, 96, 60, 90, 58]
}

df = pd.DataFrame(data)

# Create a CSV in Colab's virtual environment (temporary storage):
df.to_csv('students.csv', index=False)
```

```python
# Read 'students.csv' file
df = pd.read_csv('/content/students.csv')
# Read it into a pandas DataFrame and print the first 3 rows.
print(df.head(3))
```

```python
#Calculate Average Scores
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
print("\nStudent with highest average:")
print(df[df['Average'] == df['Average'].max()])
```

```python
#Students with Science > 80
print("\nStudents scoring more than 80 in Science:")
print(df[df['Science'] > 80])
```

```python
#Students with Average < 70
print("\nStudents with average score below 70:")
print(df[df['Average'] < 70])
```

```python
#Sorting by English(descending)
print("Top 3 students by English score:")
print(df.sort_values(by='English', ascending=False).head(3))
```

```python
#Bar chart of Average scores
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Average'], color='skyblue')
plt.title('Average Score per Student')
plt.xlabel('Student Name')
plt.ylabel('Average Score')
plt.tight_layout()
plt.savefig('average_bar_chart.png')  # optional
plt.show()
```

```python
#Line chart of all subjects
df.set_index('Name')[['Math', 'Science', 'English']].plot(kind='line', marker='o')
plt.title('Subject Scores by Student')
plt.ylabel('Scores')
plt.grid(True)
plt.tight_layout()
plt.savefig('subject_line_chart.png')  # optional
plt.show()
```

```python
#Subject with highest total score
subject_totals = df[['Math', 'Science', 'English']].sum()
print("\nSubject with highest total score:", subject_totals.idxmax())
```

```python
#Save summary as CSV
df.to_csv('students_summary.csv', index=False)
print("students_summary.csv saved successfully.")
```
